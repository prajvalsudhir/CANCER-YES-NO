```python
## train split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
## scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
# creating the network
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
model.add(Dense(30,activation='relu'))
model.add(Dense(30,activation='relu'))

# output layer will have sigmoid acti function  for classification
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
# we will overfit the data by training it without any early stopping
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600)
```

```python
pd.DataFrame(model.history.history)
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
## the validation loss is increasing beyond the training loss so we have overfit the data as its not
## able to predict new values accurately
```

```python
## early stopping callbacks and dropout layers
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# add a dropout layer between each hidden layer to stop that % of neurons from activating 
# and the weights and biases wont be updated
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

# output layer will have sigmoid acti function  for classification
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[EarlyStopping(monitor='val_loss',patience=25,mode='min')])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
# add a dropout layer between each hidden layer to stop that % of neurons from activating 
# and the weights and biases wont be updated
## upon adding a dropout layer reduce the next hidden layer by half as dropout layer only activated for half
## the previous neurons so only those outputs are given to the next layer
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

# output layer will have sigmoid acti function  for classification
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[EarlyStopping(monitor='val_loss',patience=25,mode='min')])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
print(classification_report(y_test,y_pred))
```

```python
#for binary classification predic the classes
y_pred = model.predict_classes(X_test)
```

```python
y_pred
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
